```python
from dotenv import load_dotenv
import os
load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY") #OpenAI key 가져오기
```

```python
from openai import OpenAI

client = OpenAI(api_key=OPENAI_API_KEY)
# "client" 변수에 OpenAI api key를 저장.
embedding = client.embeddings.create(
  input="내일 날씨는 어때?",
  model="text-embedding-3-small"
)
# client.embeddings.create(): api key를 이용해 "input" 변수를 "model" 변수를 사용하여 임베딩으로 변환하는 함수.
embedding
```

```python
import numpy as np

def get_embedding(text, model="text-embedding-3-small"):
  client = OpenAI(api_key=OPENAI_API_KEY)
  response = client.embeddings.create(
    input=text,
    model=model
  )
  return response.data[0].embedding
# get_embedding(): input text에 대한 임베딩을 반환하는 함수. 함수 내용은 위와 동일.

def cosine_similarity(a, b):
  return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
# cosine_similarity(a,b) = (a, b 벡터의 내적) / (a 벡터 크기)* (b 벡터 크기)

text1 = "오늘은 날씨가 좋아서 공원에 산책을 갔습니다."
text2 = "날씨가 맑아서 야외 활동하기 좋은 날이었습니다."

embedding1 = get_embedding(text1)
embedding2 = get_embedding(text2)

similarity = cosine_similarity(embedding1, embedding2)

print(f"두 텍스트의 유사도: {similarity}")
```
